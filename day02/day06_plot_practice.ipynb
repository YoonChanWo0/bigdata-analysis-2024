## Matplotlib, Seaborn활용

```python
# 필요 라이브버리 import
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
# 한글 꺠짐 문제 해결
from matplotlib import rcParams, font_manager, rc

font_path = 'C:/Windows/Fonts/NanumGothicCoding.ttf' # 나눔고딕체 사용, 나눔고딕에서 오류발생!
font = font_manager.FontProperties(fname=font_path).get_name() # 실제 설치된 폰트 이름조회
rc('font', family=font) # 한글꺠짐현상 해결!!
rcParams['axes.unicode_minus'] = False # 한글 사용시 마이너스 표시 꺠짐 해결!
```

- 타이타닉 데이터

```python
sns.get_dataset_names()
```

```python
titanics = sns.load_dataset('titanic')
```

```python
titanics.info()
```

```python
pclassSurvived = pd.crosstab(titanics['pclass'], titanics['survived'])
pclassSurvived
```

```python
pclassSurvived.columns = pclassSurvived.columns.map({0:'Dead', 1:'Alive'})
```

```python
pclassSurvived
```

```python
sns.set_style('whitegrid')
```

```python
pclassSurvived.plot.bar()
pclassSurvived.plot.bar(subplots=True)
plt.show()
```

```python
# 라인차트
pclassSurvived.plot.line()
#pclassSurvived['Dead'].plot.line()
plt.show()
```

```python
# 파이차트
pclassSurvived['Dead'].plot.pie()
plt.show()
```
