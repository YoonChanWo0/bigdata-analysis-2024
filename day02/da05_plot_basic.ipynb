```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## 단일차트
plt.plot([1,2,3,4,5])
plt.show() # [<matplotlib.lines.Line2D at 0x16d1f91cce0>] 와 같은 객체 정보 출력이 사라짐
```

```python
plt.plot(np.random.randn(50)) # np.random.randn(n) n개의 정수를 랜덤하게 만들어줌
plt.show()
```

```python
## 다중차트, figure() 사용 4개 이상 불가
fig = plt.figure()
plt1 = fig.add_subplot(2, 2, 1) # 1번 차트
plt2 = fig.add_subplot(2,2,2) # 2번 차트
plt3 = fig.add_subplot(2,2,3) # 3번 차트
plt4 = fig.add_subplot(2,2,4) # 4번 차트

plt1.plot(np.random.randn(50))
plt1.hist(np.random.randn(100))
plt3.scatter(np.arange(30),np.random.randn(30))
plt4.plot(np.random.randn(30),'violet') # 색상 'r', 'g', 'b', 'y', 'violet'

plt.show()
```

```python
## 다중차트 2번쨰
# subplots() figsize로 차트 넓이 조정가능
# color = 'k' 검은색, alpha=0.0완전투명 1.0투명도 없음
fig, axes = plt.subplots(2, 3, figsize=(18, 6))
axes[0][0].plot(np.random.randn(50), color='g', alpha=1.0)
# bins는 그루핑개수
axes[0][1].hist(np.random.randn(100), color='g', bins=30)
axes[0][2].plot(np.random.randn(50), color='b', linestyle='-')
# marker *(별), ^(삼각형)
axes[1][0].scatter(np.arange(30),np.random.randn(30), np.random.randn(30), marker='*')
axes[1][1].plot(np.random.randn(30),'violet')
axes[1][2].plot(np.random.randn(20),marker='o', label='o')
plt.show()
```

```python
# cumsum() cumulative sum(누적합)
# legend() 범례표시함수
#플롯에 데이터 여러개 표시되면 파란색, 주황색, 녹색 순
plt.plot(np.random.randn(50).cumsum(), label='data1')
plt.plot(np.random.randn(50).cumsum(), label='data2')
plt.plot(np.random.randn(50).cumsum(), label='data3')
# 위치값 
plt.legend(loc='upper right')
plt.show()
```

#### 차트 스타일 지정 옵션

```python
## x, y축 설정
plt.figure(figsize=(18, 5))
plt.plot(np.random.randn(1000))
plt.xticks([0, 250, 750, 1000], labels=['min', '25%', 'median', '75%', 'max'])
## y축도 동일하게 변경 가능
plt.title('Stastics') # 한글은 기본적 지원 안함
plt.show()
```

#### Matplotlib한글 표시 문제
- 기본적으로 한글을 지원.font_manager 객체사용해서 한글을 쓸 수 있도록 변경

- 한글을 기본으로 쓰기 떄문에 항상 이 아래부분은 노트북쓸떄마다 복사해서 사용

```python
# 한글 꺠짐 문제 해결
from matplotlib import rcParams, font_manager, rc

font_path = 'C:/Windows/Fonts/NanumGothicCoding.ttf' # 나눔고딕체 사용, 나눔고딕에서 오류발생!
font = font_manager.FontProperties(fname=font_path).get_name() # 실제 설치된 폰트 이름조회
rc('font', family=font) # 한글꺠짐현상 해결!!
rcParams['axes.unicode_minus'] = False # 한글 사용시 마이너스 표시 꺠짐 해결!
```
